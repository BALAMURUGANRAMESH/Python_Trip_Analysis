```python
import pandas as pd
import seaborn as sns
import  matplotlib.pyplot as plt
```

-- Data Cleaning and pre-processing --

The ‘Process’ phase involves transforming raw data into actionable insights through systematic steps.

With regards to the project, the specific actions that I carried out are as follows:

```python
#1 import each file individually.
jan_2025 = pd.read_csv('202501.csv')
feb_2025 = pd.read_csv('202502.csv')
mar_2025 = pd.read_csv('202503.csv')
apr_2025 = pd.read_csv('202504.csv')
may_2025 = pd.read_csv('202505.csv')
jun_2025 = pd.read_csv('202506.csv')
jul_2025 = pd.read_csv('202507.csv')
aug_2025 = pd.read_csv('202508.csv')
sep_2025 = pd.read_csv('202509.csv')
oct_2025 = pd.read_csv('202510.csv')
nov_2025 = pd.read_csv('202511.csv')
dec_2025 = pd.read_csv('202512.csv')
```

```python
#2 Merge these files into one data frame
df=pd.concat([jan_2025,feb_2025,mar_2025,apr_2025,may_2025,jun_2025,jul_2025,aug_2025,sep_2025,oct_2025,nov_2025,dec_2025])
```

```python
#3 Identify number of columns and rows.
df.shape
```

```python
#4 Identify the columns and type of each column of the data frame.
df.info(max_cols=15)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
#5 Check for null values.
df.isnull().sum() # To check the amount of null values in each column
```

```python
df.dropna(inplace = True)  # To drop(remove) the null values
```

```python
df.isnull().sum() # To recheck the amount of null values, make sure it is 0
```

```python
#6 Check for duplicated values.
df.duplicated().sum() # To check the amount of duplicated values
df.drop_duplicates() # To remove the dupliacted values
```

```python
#7 Modify the data type of certain column.
df["started_at"] = pd.to_datetime(df["started_at"],errors='coerce')
df["ended_at"] = pd.to_datetime(df["ended_at"],errors='coerce')

df.info() # To check the updated data type
```

```python
#8 Create new columns that extract the hour, date, day, and month
df['hour'] = df['started_at'].dt.hour # To extract the hour
df['date'] = df['started_at'].dt.day #To extract the date
df['day'] = df['started_at'].dt.day_name() # To extract the day of the week
df['month'] = df['started_at'].dt.month_name() # To extract the month

df.head()
```

```python
# To rearrange the days

day_ordered = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
               'Saturday', 'Sunday']

df['day'] = pd.Categorical(df['day'], categories=day_ordered, ordered=True)



# To rearrange the months
month_ordered = ['January', 'February', 'March', 'April', 'May',
                 'June', 'July', 'August', 'September', 'October',
                 'November', 'December']

df['month'] = pd.Categorical(df['month'], categories=month_ordered, ordered=True)
```

```python
#9 Create a new column that calculates the duration of each ride.
df['ride_duration'] = (df['ended_at'] - df['started_at'])/ pd.Timedelta(minutes=1)
#Express the ride duration in terms of only minutes , and not minutes and seconds

df.head()
```

```python
#10 Verify whether there were negative values for the ride duration, which would be illogical.
df[df['ride_duration'] < 0].count()
```

```python
df = df[df['ride_duration'] > 0] # Remove the rows with negative ride durations
df[df['ride_duration'] < 0].count() # To reconfirm there are no longer negative ride durations
```

```python
#11 Export the cleaned dataset as a .csv file.
df.to_csv("2025_cleaned.csv")
```

-- Analyze & Share --

Exploratory Data Analysis (EDA) is a critical, iterative process used by data scientists to analyze, visualize, and summarize datasets to uncover patterns, spot anomalies, test hypotheses, and check assumptions before formal modeling.

I did my analysis using Python Pandas, and selected MatPlotlib Express for the data visualizations. The questions that I explored were:

```python
#Analysis of the number of casual and membered riders.
#Number of rides taken by casual and membered riders.
Number_of_riders = df['member_casual'].value_counts()
Number_of_riders
```

```python
#Percentage of casual and membered riders.
plt.figure(figsize=(6, 3))
colors = ['#ff9999','#66b3ff'] # Soft red and blue

plt.pie(
    Number_of_riders,
    labels=Number_of_riders.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'white'}
)
plt.legend(title="Rider Type", loc="center right",bbox_to_anchor=(1.5, 0.5))
plt.title('Percentage of Casual vs. Member Riders')
plt.show()
```

```python
#Analysis of the number of rides.
#Number of rides taken throughout different hours.
rides_hour = df.groupby(['member_casual','hour'], as_index=False)['ride_id'].count()
# Pivot data for Matplotlib
pivot_df = rides_hour.pivot(index='hour', columns='member_casual', values='ride_id')

# Plot
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 3), color=['Orange','Purple'])

plt.title('Numbers of Rides by Member Types Across Hours')
plt.xlabel('Hours of the Day')
plt.ylabel('Number of Rides')
plt.show()
```

```python
#Number of rides taken throughout different days of the week.
rides_day = df.groupby(['member_casual','day'],as_index=False,observed=False)['ride_id'].count()
#rides_day
# Pivot data for Matplotlib
pivot_df = rides_day.pivot(index='day', columns='member_casual', values='ride_id')

# Plot
pivot_df.plot(kind='line', stacked=False, marker='o', figsize=(10, 3))

plt.title('Numbers of Rides by Member Types Across Days')
plt.xlabel('Days of the week (2025)')
plt.ylabel('Number of Rides')
plt.show()
```

```python
#Number of rides taken throughout different months.
rides_month = df.groupby(['member_casual','month'], as_index=False,observed=False)['ride_id'].count()
# Pivot data for Matplotlib
pivot_df = rides_month.pivot(index='month', columns='member_casual', values='ride_id')

# Plot
#ax=pivot_df.plot(kind='bar', stacked=True, figsize=(10, 3), color=['Red','Purple'])
ax=pivot_df.plot.bar(rot=0, figsize=(10, 3))

for container in ax.containers:
    custom_labels = [f'{val/1000:,.0f}K' for val in container.datavalues]
    ax.bar_label(container, labels= custom_labels, label_type='edge', fontsize=9, color='black', rotation=90)

plt.title('Numbers of Rides by Member Types Across Month')
plt.xlabel('Months of 2025')
plt.ylabel('Number of Rides')
plt.show()
```

```python
#Analysis of the ride durations.
#3.1 Ride durations throughout different days of the week.
rides_day = df.groupby(['member_casual','day'],as_index=False,observed=False)['ride_duration'].count()
#rides_day
# Pivot data for Matplotlib
pivot_df = rides_day.pivot(index='day', columns='member_casual', values='ride_duration')

# Plot
ax=pivot_df.plot.line(marker='x', figsize=(10, 3))

# Iterate through each column (Member/Casual) to label points
for column in pivot_df.columns:
    for i, val in enumerate(pivot_df[column]):
        # Format the label (e.g., 1500 -> 2K)
        label = f'{val/1000:,.1f}K'

        # Annotate each point
        ax.annotate(label,
                    xy=(i, val),
                    xytext=(0, 7), # 7 points vertical offset
                    textcoords="offset points",
                    ha='center',
                    fontsize=9,
                    rotation=0)

plt.title('Ride durations by Member Types Across Days')
plt.xlabel('Days of the week (2025)')
plt.ylabel('Durations of Rides (Mintues)')
plt.show()
```

```python
#Analysis of the types of bikes.
#Which type of bike is most popular?
bikes_popularirty_day = df.groupby(['member_casual','rideable_type', 'day'], as_index = False, observed=False)['ride_id'].count()
# Use pivot_table instead of pivot to handle the duplicates by summing them
pivot_df = bikes_popularirty_day.pivot_table(index='day', columns=['rideable_type', 'member_casual'], values='ride_id', aggfunc='sum', observed=False)

ax=pivot_df.plot(kind='bar', stacked=True, figsize=(10, 3))

for container in ax.containers:
    custom_labels = [f'{val/1000:,.0f}K' for val in container.datavalues]
    ax.bar_label(container, labels= custom_labels, label_type='center', fontsize=9, color='black', rotation=90)

plt.title('Popularity of Bike Across Days (2025)')
plt.legend(title='Bike Type / User Type',bbox_to_anchor=(1.0, 1.0))
plt.xlabel('Types of Bikes')
plt.ylabel('Number of Rides')
plt.show()
```

```python
#Popularity of bikes without considering type of riders.

bike_general = df.groupby(['rideable_type','day'], as_index=False,observed=False)['ride_id'].count()
# Pivot data for Matplotlib
pivot_df = bike_general.pivot(index='day', columns='rideable_type', values='ride_id')

# Plot
#ax=pivot_df.plot.bar(kind='bar', stacked=True, figsize=(10, 3), color=['Red','Purple'])
ax=pivot_df.plot.bar(rot=0, figsize=(10, 3))

for container in ax.containers:
    custom_labels = [f'{val/1000:,.0f}K' for val in container.datavalues]
    ax.bar_label(container, labels= custom_labels, label_type='center', fontsize=9, color='black', rotation=90)

plt.title('Popularity of Bike Across Days Without Considering Membership (2025)')
plt.legend(title='Bike Type', bbox_to_anchor=(1.0, 1.0))
plt.xlabel('Types of Bikes')
plt.ylabel('Number of Rides')
plt.show()
```
